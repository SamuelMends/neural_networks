```python
from sklearn.neural_network import MLPClassifier
from sklearn import datasets
```

```python
iris = datasets.load_iris()
entradas = iris.data
saidas = iris.target
```

```python
redeNeural = MLPClassifier(verbose=True,
                           max_iter=1000,
                           tol=0.00001,
                           activation='logistic')
redeNeural.fit(entradas,saidas)
```
