```python
import numpy as np 
```

```python
def sigmoide(soma):
    return 1 / (1 + np.exp(-soma))
```

```python
def sigmoideDerivada(sig):
    return sig * (1 - sig)
```

```python
entradas = np.array([[0,0],[0,1],[1,0],[1,1]])
saidas = np.array([[0],[1],[1],[0]])
```

```python
peso0 = 2*np.random.random((2,3)) - 1 
peso1 = 2*np.random.random((3,1)) - 1
```

```python
epocas = 100000
taxaAprendizagem = 0.6
momento = 1
```

```python
for j in range(epocas):
    camadaEntradas = entradas
    somasinapse0 = np.dot(camadaEntradas, peso0)
    camadaOCulta = sigmoide(somasinapse0)
    somasinapse1 = np.dot(camadaOCulta, peso1)
    camadasaida = sigmoide(somasinapse1)
    errocamadasaida = saidas - camadasaida
    mediabsoluta = np.mean(np.abs(errocamadasaida))
    print("Erro: " + str(mediabsoluta))
    derivadaSaida = sigmoideDerivada(camadasaida)
    deltaSaida = errocamadasaida * derivadaSaida
    peso1Transporta = peso1.T
    deltaSaidaXPeso = deltaSaida.dot(peso1Transporta)
    deltaCamadaOculta = deltaSaidaXPeso * sigmoideDerivada(camadaOCulta)
    camadaOCultaTransposta = camadaOCulta.T
    pesoNovo1 = camadaOCultaTransposta.dot(deltaSaida)
    peso1 = (peso1 * momento) + (pesoNovo1 * taxaAprendizagem)
    camadaEntradaTransposta = camadaEntradas.T
    pesoNovo0 = camadaEntradaTransposta.dot(deltaCamadaOculta)
    peso0 = (peso0 * momento) + (pesoNovo0 * taxaAprendizagem)
    
```
